# Import the desired Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Load the data 

```python
df = pd.read_csv('credit_risk_dataset.csv')

df.head()
```

```python
df.columns
```

# checking the column names and keeping the target vairable at last column

```python
df=df[['person_age', 'person_income', 'person_home_ownership',
       'person_emp_length', 'loan_intent', 'loan_grade', 'loan_amnt',
       'loan_int_rate', 'loan_percent_income',
       'cb_person_default_on_file', 'cb_person_cred_hist_length','loan_status']]
```

```python
df
```

# Checking the Null values and removing them

```python
df.isnull().sum()
```

```python
fdf=df.dropna(axis=0, how='any')
```

```python
fdf
```

# Removing the extreme outliers in age and employement length

```python
odf=fdf[fdf.person_age<=95]
    
    
```

```python
credit=odf[odf.person_emp_length<=40]
```

```python
credit
```

# Start of SKlearn

```python
import sklearn

print(sklearn.__version__)
```

```python
x=credit[['person_age', 'person_income', 'person_home_ownership',
       'person_emp_length', 'loan_intent', 'loan_grade', 'loan_amnt',
       'loan_int_rate', 'loan_percent_income',
       'cb_person_default_on_file', 'cb_person_cred_hist_length']]
y=credit['loan_status']
```

# split into train and test

```python


from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, train_size=0.75, random_state=100)
```

```python
X_train_num = X_train.select_dtypes(include=['int64', 'float64'])

X_train_num.head()
```

# scaling the numerical features

```python

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_num_rescaled = pd.DataFrame(scaler.fit_transform(X_train_num), 
                                    columns = X_train_num.columns, 
                                    index = X_train_num.index)

X_train_num_rescaled.head()
```

# Applying OHE and Labelling(categorical to numerical)

```python
X_train_cat1=X_train_cat[["cb_person_default_on_file","loan_intent","person_home_ownership"]]
```

```python
X_train_cat2=X_train_cat['loan_grade']
```

```python
X_train_cat1
```

```python
X_train_cat3=pd.DataFrame(X_train_cat2)
```

# OneHotEncoding the categorical features


```python

from sklearn.preprocessing import OneHotEncoder

encoder_= OneHotEncoder(drop='first', sparse=False)


X_train_cat_ohe = pd.DataFrame(encoder_.fit_transform(X_train_cat1), 
                               columns=encoder_.get_feature_names_out(X_train_cat1.columns), 
                               index = X_train_cat1.index)

X_train_cat_ohe.head()
```

# label-encoding

```python
X_train_cat3.loan_grade.unique()
```

```python
X_train_cat3.head()
```

```python
encoder2 = {'G' : 1, 'F' : 2, 'E' : 3, 'D' : 4, 'C' : 5,'B' : 6,'A' : 7}

X_train_cat_ohe['loan_grade'] = X_train_cat3['loan_grade'].apply(lambda x : encoder2[x])

X_train_cat_ohe.head()
```

# Concatination of transformed numerical and categorical.

```python
X_train_transformed = pd.concat([X_train_num_rescaled, X_train_cat_ohe], axis=1)

X_train_transformed.head()
```
