# Import the desired Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Load the data

```python
df = pd.read_csv('churn_dataset.csv')
df
```

```python
df.columns
```

```python
df.head()
```

# Deleting customer ID

```python
del df['customerID']
```

```python
df.info()
```


# deleting spaces and removing null values

```python
df = df.replace(r'^\s*$', np.nan, regex=True)
```

```python
df['TotalCharges'] = df['TotalCharges'].astype(float)
```

```python
df['tenure'] = df['tenure'].astype(float)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df=df.dropna(axis=0, how='any')
```

```python
df.isnull().sum()
```

# import sklearn

```python
import sklearn

print(sklearn.__version__)
```

```python
x=df[['gender', 'SeniorCitizen', 'Partner', 'Dependents',
       'tenure', 'PhoneService', 'MultipleLines', 'InternetService',
       'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport',
       'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling',
       'PaymentMethod', 'MonthlyCharges', 'TotalCharges']]
y=df['Churn']
```

# split into train and test

```python


from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, train_size=0.75, random_state=100)
```

```python
X_train.head()
```

```python
print(X_train.shape, y_train.shape)

print(X_test.shape, y_test.shape)
```

```python
X_train.dtypes
```

```python
X_train['SeniorCitizen'].value_counts(normalize=True)
```

# Split the dataset

```python
X_train_cat = X_train.select_dtypes(include=['object','int64'])

X_train_cat.head()
```

```python
X_train_num = X_train.select_dtypes(include=['float64'])

X_train_num.head()
```

# scaling the numerical features

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_num_rescaled = pd.DataFrame(scaler.fit_transform(X_train_num), 
                                    columns = X_train_num.columns, 
                                    index = X_train_num.index)

X_train_num_rescaled.head()
```

# Applying One hot encoding(categorical to numerical)

```python
from sklearn.preprocessing import OneHotEncoder

encoder_= OneHotEncoder(drop='first', sparse=False)


X_train_cat_ohe = pd.DataFrame(encoder_.fit_transform(X_train_cat), 
                               columns=encoder_.get_feature_names_out(X_train_cat.columns), 
                               index = X_train_cat.index)

X_train_cat_ohe.head()
```

# Concatination

```python
X_train_transformed = pd.concat([X_train_num_rescaled, X_train_cat_ohe], axis=1)

X_train_transformed.head()
```

# TEST- data pre-processing

```python
X_test.head()
```

# split based on datatypes

```python
X_test_cat = X_test.select_dtypes(include=['object','int'])

X_test_cat.head()
```

```python
X_test_transformed = pd.concat([X_test_num_rescaled, X_test_cat_ohe], axis=1)

X_test_transformed.head()
```
